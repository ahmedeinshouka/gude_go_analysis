```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='سن الأب', data=data, order=data['سن الأب'].value_counts().index.sort_values())
plt.xlabel("Father's Age")
plt.title("Count of Respondents by Father's Age")
plt.show()
```
